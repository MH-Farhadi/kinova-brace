```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function for grouped side-by-side violin plots with subplots
def plot_side_by_side_violinplots(df, question_groups, group_titles):
    # Create a single figure for all subplots
    fig, axes = plt.subplots(len(question_groups), 2, figsize=(14, 5 * len(question_groups)))

    # Store unique legend handles and labels
    legend_items = {}

    # Iterate through each group and plot the violins
    for i, (question_group, group_title) in enumerate(zip(question_groups, group_titles)):
        # Filter the DataFrame for the specific questions
        filtered_df = df[df["Question"].isin(question_group)]

        # Create subsets for comparison
        df_aras_manual = filtered_df[filtered_df["Mode"].isin(["ARAS", "Manual"])]
        df_aras_autonomous = filtered_df[filtered_df["Mode"].isin(["ARAS", "Autonomous"])]

        # Plot ARAS vs Manual
        sns.violinplot(
            ax=axes[i, 0],
            x="Question",
            y="Score",
            hue="Mode",
            data=df_aras_manual,
            split=True,
            inner="stick",
            palette={"Manual": "orange", "ARAS": "limegreen"},
            bw_method=0.25,
            linewidth=1.2
        )
        axes[i, 0].set_ylabel("Scores (1-5)", fontsize=14)
        axes[i, 0].tick_params(axis='x', rotation=0, labelsize=14)
        axes[i, 0].tick_params(axis='y', labelsize=14)
        axes[i, 0].set_xlabel(None)

        # Collect legend items
        for handle, label in zip(*axes[i, 0].get_legend_handles_labels()):
            legend_items[label] = handle
        axes[i, 0].legend_.remove()

        # Plot ARAS vs Autonomous
        sns.violinplot(
            ax=axes[i, 1],
            x="Question",
            y="Score",
            hue="Mode",
            data=df_aras_autonomous,
            split=True,
            inner="stick",
            palette={"Autonomous": "cornflowerblue", "ARAS": "limegreen"},
            bw_method=0.25,
            linewidth=1.2
        )
        axes[i, 1].set_ylabel("Scores (1-5)", fontsize=14)
        axes[i, 1].tick_params(axis='x', rotation=0, labelsize=14)
        axes[i, 1].tick_params(axis='y', labelsize=14)
        axes[i, 1].set_xlabel(None)
        axes[i, 1].set_ylabel(None)

        # Collect legend items
        for handle, label in zip(*axes[i, 1].get_legend_handles_labels()):
            legend_items[label] = handle
        axes[i, 1].legend_.remove()

    # Adjust layout to fit subplots
    fig.tight_layout()

    # Create a single legend for the figure
    legend_order = ["Manual", "Autonomous", "ARAS"]  # Desired order of legend items
    reordered_handles_labels = [(legend_items[label], label) for label in legend_order if label in legend_items]
    reordered_handles, reordered_labels = zip(*reordered_handles_labels)

    fig.legend(
        reordered_handles,
        reordered_labels,
        loc='upper center',
        bbox_to_anchor=(0.5, 1.05),
        ncol=len(reordered_labels),
        fontsize=12,
        title="Mode",
        title_fontsize=14
    )

    plt.show()

# Question groups and titles
question_groups = [
    ["Easy", "Effort", "Flexibility", "Confidence"],
    ["Satisfaction", "Future Use"]
]
group_titles = [
    "Easy, Effort, Flexibility, Confidence",
    "Satisfaction, Future Use"
]

# Call the function to plot the grouped side-by-side violin plots
plot_side_by_side_violinplots(df, question_groups, group_titles)
```

## Spider

```python
import matplotlib.pyplot as plt
from math import pi, degrees
import numpy as np

# Data aggregation function
def calculate_averages(data):
    """Calculate the averages for each question across participants."""
    return {key: np.mean(values) for key, values in data.items()}

def plot_spider_chart(questions, modes, data_manual, data_aras, data_autonomous):
    # Calculate averages for each mode
    averages = {
        "Manual": calculate_averages(data_manual),
        "Autonomous": calculate_averages(data_autonomous),
        "ARAS": calculate_averages(data_aras),
    }

    # Define custom colors for each mode
    mode_colors = {
        "Manual": "orange",
        "Autonomous": "cornflowerblue",
        "ARAS": "limegreen",
    }

    # Number of variables (questions)
    N = len(questions)

    # Calculate angles for the radar chart
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Close the circle

    # Create the spider chart
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    # Add radial lines from the origin to the nodes
    for angle in angles[:-1]:  # Skip the duplicate angle at the end
        ax.plot([angle, angle], [0, 5], color="black", linewidth=0.8, linestyle="--")

    # Add polygonal levels with numeric labels
    for level in range(1, 6):  # Levels from 1 to 5
        level_points = [level] * len(angles)
        ax.plot(
            angles,
            level_points,
            color="black",
            linewidth=0.8,
            linestyle="--",
            alpha=0.7
        )
        # Add labels to each level
        ax.text(
            pi / 5,  # Position at the top (angle = pi/2)
            level,
            str(level),
            verticalalignment="center",
            fontsize=14
        )

    # Plot data for each mode
    for mode_name, mode_data in averages.items():
        values = [mode_data[question] for question in questions]
        values += values[:1]  # Close the data path
        ax.plot(angles, values, linewidth=2, linestyle="solid", label=mode_name, color=mode_colors[mode_name])
        ax.fill(angles, values, alpha=0.25, color=mode_colors[mode_name])

    # Add category labels
    for angle, question in zip(angles[:-1], questions):
        # Calculate the rotation angle for the label
        rotation = degrees(angle)
        if angle > pi:  # Adjust rotation for bottom half
            rotation -= 270
        else:
            rotation -= 90
        ax.text(
            angle,
            5.3,  # Position slightly outside the maximum range
            question,
            horizontalalignment="center",
            verticalalignment="center",
            fontsize=12,
            fontweight="bold",
            rotation=rotation
        )

    # Customize y-axis ticks
    ax.yaxis.set_visible(False)  # Hide y-axis labels for cleaner appearance
    plt.ylim(0, 5)  # Set the y-axis to match your score range

    # Remove angular labels while keeping the radial lines
    ax.xaxis.set_tick_params(width=0)
    ax.set_xticklabels([])

    # Add legend
    plt.legend(
        loc="upper right",
        bbox_to_anchor=(1.2, 1.1),
        fontsize=12,
        title="Modes",
        title_fontsize=14
    )

    # Add grid and adjust layout
    ax.grid(False)  # Disable the default grid lines
    plt.tight_layout()
    plt.show()

# Call the function with your data
plot_spider_chart(
    questions=[
        "Satisfaction",
        "Easy",
        "Effort",
        "Confidence",
        "Flexibility",
        "Assistance",
        "Future Use"
    ],
    modes=["Manual", "Autonomous", "ARAS"],
    data_manual=data_Manual,
    data_aras=data_ARAS,
    data_autonomous=data_Autonomous
)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def perform_wilcoxon_test(data1, data2):
    # Wilcoxon signed-rank test for ordinal paired data
    statistic, p_value = stats.wilcoxon(data1, data2)
    return p_value

def get_significance_symbol(p_value):
    if p_value > 0.05:
        return 'ns'
    if p_value < 0.0001:
        return '****'
    if p_value < 0.001:
        return '***'
    elif p_value < 0.01:
        return '**'
    elif p_value < 0.05:
        return '*'
    else:
        return ''

# Calculate means and standard errors for assistance levels
def calculate_mean_and_error(data):
    mean = np.mean(data)
    error = np.std(data) / np.sqrt(len(data))
    return mean, error

# Calculate values for each mode
mean_manual, error_manual = calculate_mean_and_error(data_Manual["Assistance"])
mean_aras, error_aras = calculate_mean_and_error(data_ARAS["Assistance"])
mean_autonomous, error_autonomous = calculate_mean_and_error(data_Autonomous["Assistance"])
mean_preferred, error_preferred = calculate_mean_and_error(data_preferred_assistance)

# Data for plotting
modes = ["Manual", "Autonomous", "ARAS", "Users' preference"]
means = [mean_manual, mean_autonomous, mean_aras, mean_preferred]
errors = [error_manual, error_autonomous, error_aras, error_preferred]

# Perform Wilcoxon signed-rank tests and store results
preference_index = 3
data_arrays = [
    data_Manual["Assistance"],
    data_Autonomous["Assistance"],
    data_ARAS["Assistance"],
    data_preferred_assistance
]
test_results = []
for i in range(len(modes) - 1):
    p_value = perform_wilcoxon_test(data_arrays[i], data_preferred_assistance)
    stars = get_significance_symbol(p_value)
    if stars:
        test_results.append((i, p_value, stars))

# Sort test results to get ARAS (2) first, then Autonomous (1), then Manual (0)
desired_order = [2, 1, 0]  # ARAS, Autonomous, Manual
significance_data = []
for idx in desired_order:
    for result in test_results:
        if result[0] == idx:
            significance_data.append((result[0], preference_index, result[2]))

# Plot
plt.figure(figsize=(6, 5))
x = np.arange(len(modes)) * 1.4
bars = plt.bar(x, means, yerr=errors, capsize=5,
               color=["orange", "cornflowerblue", "limegreen", "gray"],
               alpha=0.8)

# NEW LINE: apply hatching only to the blue bar (index 1)
bars[1].set_hatch('//')

# Add significance brackets in specified order
max_height = max(means) + max(errors)
bracket_height = max_height + 0
height_increment = 0.4

for i, (group1, group2, stars) in enumerate(significance_data):
    # Calculate positions for the bracket
    x1 = x[group1]
    x2 = x[group2]
    current_height = bracket_height + (i * height_increment)

    # Draw the bracket
    plt.plot([x1, x1, x2, x2],
             [current_height - 0.1, current_height, current_height, current_height - 0.1],
             color='black', linewidth=1.2)

    # Add stars
    plt.text((x1 + x2) / 2, current_height, stars,
             ha='center', va='bottom', fontsize=18, fontweight='bold')

# Formatting
plt.xticks(x, modes, fontsize=14)

# Customize y-axis ticks
plt.yticks(ticks=np.arange(0, 6, 1), labels=[str(i) for i in range(6)], fontsize=16)
plt.ylabel("Scores", fontsize=18)
plt.ylim(0, 7)  # Keep the gap for significance stars without showing numbers above 5

plt.ylabel("Assistance Level (1-5)", fontsize=18)
plt.ylim(0, bracket_height + len(significance_data) * height_increment + 0.2)

# Remove top and right spines (borders)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

# Show plot
plt.show()
```

```python
import cv2
import numpy as np

# Define the file path
file_path = "/content/4.txt"

# Define the action mapping
action_map = {'a': 'left', 'd': 'right', 'x': 'neutral'}

# Read the file and store actions in a list with their correct names
actions = []
with open(file_path, "r") as file:
    for line in file:
        char = line.strip()
        if char not in action_map:  # Replace any other character with 'x'
            char = 'x'
        actions.append(action_map[char])

# Video settings
frame_width, frame_height = 600, 800  # Screen size
frame_rate = 2  # 500ms per frame
output_path = "inputs_4.mp4"

# Text settings
line_height = 40  # Space between lines
max_lines = (frame_height - 50) // line_height  # Ensures last line is fully visible
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (0, 0, 0)
thickness = 2

# Initialize video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (frame_width, frame_height))

# Generate frames with scrolling effect
displayed_actions = []
for action in actions:
    if len(displayed_actions) >= max_lines:
        displayed_actions.pop(0)  # Remove the topmost action to make space
    displayed_actions.append(action)

    # Create a new blank frame
    frame = np.ones((frame_height, frame_width, 3), dtype=np.uint8) * 255  # White background

    # Render text for displayed actions
    for idx, act in enumerate(displayed_actions):
        text_position = (50, 50 + idx * line_height)
        cv2.putText(frame, act, text_position, font, font_scale, font_color, thickness)

    # Write frame to video
    out.write(frame)

# Release video writer
out.release()

print(f"Scrolling video saved to {output_path}")
```
